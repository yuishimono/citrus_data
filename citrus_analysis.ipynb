```python
#Pandasのインポート
import pandas as pd
#numpyのインポート
import numpy as np
#reのインポート
import re
```

```python
#フォルダやファイルパスの定義
data_dir='../data' #このファイルから見たdataフォルダのパス
input_dir=f'{data_dir}/input' #変数data_dirから見たinputフォルダのパス
output_dir=f'{data_dir}/output'
input_file=f'{input_dir}/original_data.csv'
output_file=f'{output_dir}/citrus_data.csv'
```

```python
#データの読み込み
df = pd.read_csv(input_file)
```

```python
#データの表示
display (df)
```

```python
#空白をNaNに変換する
df=df.replace(r'^\s+$', np.nan, regex=True)
#品目名の列について上のセルを参照して空白セルをうめる
df['品目名'] = df['品目名'].ffill()

display(df)
```

```python
#文字列を数値に変換する
df=df.replace(',','', regex=True) #カンマを取り除く
df['栽培面積'] = pd.to_numeric(df['栽培面積'], errors="coerce")
df['収穫量'] = pd.to_numeric(df['収穫量'], errors="coerce")
df['出荷量'] = pd.to_numeric(df['出荷量'], errors="coerce")
df['うち加工向け'] = pd.to_numeric(df['うち加工向け'], errors="coerce")
df.dtypes
```

```python
#年次を西暦に変換する

#元年を1年に変換
df['年次']=df['年次'].str.replace('元','1')

#和暦を西暦に変換
df['年次2']= df['年次'].str.replace('\d','') #元号を列「年次2」に抽出
df=df.replace(r'^\s*$', np.nan, regex=True) #列「年次2」について空白セルをNaNに変換
df['年次2'] = df['年次2'].ffill() #列「年次2」について上のセルを参照して空白セルをうめる
df['年次2']= df['年次2'].str.replace('S','1925') #列「年次2」について元号を数字に変換
df['年次2']= df['年次2'].str.replace('H','1988')
df['年次2']= df['年次2'].str.replace('R','2018')
df['年次3'] = df['年次'].str.extract('(\d+)', expand=True) #数字（和暦）を抽出
df['年次2'] = pd.to_numeric(df['年次2'], errors="coerce") #数値に変換
df['年次3'] = pd.to_numeric(df['年次3'], errors="coerce")
df['年次'] = df['年次2']+df['年次3'] #たし算の結果を年次に書き換える
display(df)
```

```python
df = df.drop(columns=df.columns[[8,9,10,11,12,13,14]])
display (df)
```

```python
#csvへの書き出し
df.to_csv(output_file,index=False)
```
